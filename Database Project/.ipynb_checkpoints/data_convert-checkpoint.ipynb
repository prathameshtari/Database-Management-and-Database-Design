```python

import numpy as np 
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns

database ='database_final.sqlite'
```

```python
con = sqlite3.connect(database)

tables = pd.read_sql("""SELECT *
                        FROM sqlite_master
                        WHERE type='table';""", con)
tables
```

A database like MySQL or PostgreSQL is considered a RDBMS (relational database management system) it's all in the name. You can through a normalised database structure of separate tables, join one or more tables together. 

In NoSQL the premise of joins is nonexistent, if you want to have relational data in a document database the general approach is to use embedded documents (sub-arrays within a key inside of a document) or reference objects of other documents.

We use joins in our relational database to convert and store the resulting dataset into a csv format

```python
match = pd.read_sql("""SELECT * FROM Match""",con)
match
```

```python
matches_all = pd.read_sql("""SELECT Match.id, Country.name AS country_name,League.name AS league_name, season, stage, date,
            HT.team_long_name AS  home_team, AT.team_long_name AS away_team, home_team_goal, away_team_goal,
            
            H1.player_name AS home_player_1,H2.player_name AS home_player_2,H3.player_name AS home_player_3,
            H4.player_name AS home_player_4,H5.player_name AS home_player_5,H6.player_name AS home_player_6,
            H7.player_name AS home_player_7,H8.player_name AS home_player_8,H9.player_name AS home_player_9,
            H10.player_name AS home_player_10,H11.player_name AS home_player_11,
            
            A1.player_name AS away_player_1,A2.player_name AS away_player_2,A3.player_name AS away_player_3,
            A4.player_name AS away_player_4,A5.player_name AS away_player_5,A6.player_name AS away_player_6,
            A7.player_name AS away_player_7,A8.player_name AS away_player_8,A9.player_name AS away_player_9,
            A10.player_name AS away_player_10,A11.player_name AS away_player_11
                                
                                
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
                                LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
                                
                                INNER JOIN Player AS H1 ON Match.home_player_1 = H1.player_api_id
                                INNER JOIN Player AS H2 ON Match.home_player_2 = H2.player_api_id
                                INNER JOIN Player AS H3 ON Match.home_player_3 = H3.player_api_id
                                INNER JOIN Player AS H4 ON Match.home_player_4 = H4.player_api_id
                                INNER JOIN Player AS H5 ON Match.home_player_5 = H5.player_api_id
                                INNER JOIN Player AS H6 ON Match.home_player_6 = H6.player_api_id
                                INNER JOIN Player AS H7 ON Match.home_player_7 = H7.player_api_id
                                INNER JOIN Player AS H8 ON Match.home_player_8 = H8.player_api_id
                                INNER JOIN Player AS H9 ON Match.home_player_9 = H9.player_api_id
                                INNER JOIN Player AS H10 ON Match.home_player_10 = H10.player_api_id
                                INNER JOIN Player AS H11 ON Match.home_player_11 = H11.player_api_id
                                
                                INNER JOIN Player AS A1 ON Match.away_player_1 = A1.player_api_id
                                INNER JOIN Player AS A2 ON Match.away_player_2 = A2.player_api_id
                                INNER JOIN Player AS A3 ON Match.away_player_3 = A3.player_api_id
                                INNER JOIN Player AS A4 ON Match.away_player_4 = A4.player_api_id
                                INNER JOIN Player AS A5 ON Match.away_player_5 = A5.player_api_id
                                INNER JOIN Player AS A6 ON Match.away_player_6 = A6.player_api_id
                                INNER JOIN Player AS A7 ON Match.away_player_7 = A7.player_api_id
                                INNER JOIN Player AS A8 ON Match.away_player_8 = A8.player_api_id
                                INNER JOIN Player AS A9 ON Match.away_player_9 = A9.player_api_id
                                INNER JOIN Player AS A10 ON Match.away_player_10 = A10.player_api_id
                                INNER JOIN Player AS A11 ON Match.away_player_11 = A11.player_api_id
                                                    
                                ORDER by date""",con)

matches_all
```

```python
matches_all.to_csv('non_rel_match.csv')
```
