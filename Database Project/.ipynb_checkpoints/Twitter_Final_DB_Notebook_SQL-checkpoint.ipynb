Since it is a relational database and we are implementing normalization we will be mapping the information collected from the twittter api to the primary key of the entity table. Here we are using twitter api to get data regarding the entity players so we are using player_api_id from players table in the database as a key to create our table

```python
#213501morata#107417#hazard#94086Willian#170323Courtois#319300kante#52133luiz#fabregas#bak#cahill
player_id = 23783
q ="cahill"
q
```

In the below query we insert the data related to the player in the database.
Running this query again creates duplicates in the database.

```python
for tweet in tweepy.Cursor(api.search, q=('"%23cahill'),since='2018-04-22', until='2018-04-23').items(20):
    
        Name = tweet.author.name
        Screen_name = tweet.author.screen_name
        Twee_created = tweet.created_at
        Retweeted = tweet.retweeted
        Favourited = tweet.favorited
        Location = tweet.user.location
        Time_zone = tweet.user.time_zone
        Geo = tweet.geo
        Text = tweet.text
        Text=Text.replace('"','')
        Favourites_Count = tweet.user.favourites_count
        Friends_Count = tweet.user.friends_count
        Followers_Count = tweet.user.followers_count
        Statuses_Count = tweet.user.statuses_count
        try:
                Hashtags = tweet.entities['hashtags'][0]['text']
        except IndexError:
                Hashtags = "None"
        URL = tweet.entities['urls']

        
        format_str = """INSERT INTO TWITTER (Player_id, User_Name, Screen_Name, Tweet_created, Retweeted, Favourited, Location, 
                        Time_Zone, Geo, Text, Favourites_Count, Friends_Count, Followers_Count, Statuses_Count, Hashtags, URL)
                        VALUES("{id}", "{user}","{screen}","{created}","{retweet}","{fav}","{loc}","{zone}","{geo}","{text}",
                                "{fav_count}","{friends}","{follow_count}","{status_count}","{hashtags}","{url}");"""
    
        sql_command = format_str.format(id=player_id, user=Name, screen=Screen_name, created=Twee_created, retweet=Retweeted, 
                                       fav=Favourited, loc=Location, zone=Time_zone, geo=Geo, text=Text, fav_count=Favourites_Count,
                                       friends=Friends_Count, follow_count=Followers_Count, status_count=Statuses_Count,
                                        hashtags=Hashtags, url=URL)
        cursor.execute(sql_command)
        
        con.commit()
```

### Question 2: How viral are my posts?

For checking how viral someones posts are we just take the retweet_count for that user. 

```python
pd.read_sql("select retweet_count from twitter where user_name='Lynx';",con)
```

### Question 3. How much influence to my posts have?

For checking how much influence my posts have we take the sum of favourites_count, friends_count, followers_count and retweet_count. 

```python
pd.read_sql("select favourites_count+friends_count+followers_count+retweet_count AS Influence from twitter where user_name='Lynx';",con)
```

### Question 6: Who should i be following?

We have a field named followers_count in our database.
So we should be following someone with a high follower_count

```python
pd.read_sql("select user_name,followers_count from twitter order by followers_count DESC;",con)
```

### Question 7: What topics are trending in my domain?

We have 2 columns in our database named 'tweet_created' and 'retweet_count'. So for trending tweets we can check the tweet_created column and retweet_count gives us the most popular tweets.

```python
pd.read_sql("select player_id, user_name, retweet_count from twitter where tweet_created > '2018-04-22 22:42:53' and retweet_count>5000;",con)
```

### Question 8: What keywords/hashtags should i add to my post?

We order the retweet_count in descending order. Then check the hashtags used in the top retweets. These give us the keywords/hashtags we should add to the post.

```python
pd.read_sql("select retweet_count, hashtags from twitter order by retweet_count desc;",con)
```

### Question 9: Should i follow somebody back?

We take the sum of followers_count, retweet_count and friends_count.
So we should follow somebody back if he has high count.

```python
pd.read_sql("select user_name, followers_count + retweet_count + friends_count from twitter order by retweet_count DESC;",con)
```

### Question 10: What is the best time to post?
